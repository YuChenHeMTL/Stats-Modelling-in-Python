```python
import requests
```

```python
res = requests.get('http://api.citybik.es/v2/networks').json()

# Get the list of cities
city_dict = dict([(city['location']['city'], city['href']) for city in res['networks']])
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
import re, json

match = [val for key, val in city_dict.items() if re.search('Montréal', key)][0]
res = requests.get('http://api.citybik.es/' + match).json()['network']['stations']
```
